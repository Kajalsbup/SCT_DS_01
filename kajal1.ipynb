```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Sample data`
data = {
'Age Group': ['0-14', '15-24', '25-34', '35-44', '45-54', '55-64', '65+'],
'Population': [331002651, 37742154, 83783942, 67081000, 65273511, 42111111, 21111111]
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
df.shape
```

```python
df.info()
```

```python
# Create a bar plot with different colors for each bar
plt.figure(figsize=(12, 6))
plt.bar(df['Age Group'], df['Population'], color=['red', 'green', 'blue', 'yellow', 'orange', 'purple', 'pink'])
plt.xlabel('Age Group')
plt.ylabel('Population')
plt.title('Bar Chart for Population Distribution by Age Group')
plt.tight_layout() # Ensure labels fit within the figure
plt.show()
```
